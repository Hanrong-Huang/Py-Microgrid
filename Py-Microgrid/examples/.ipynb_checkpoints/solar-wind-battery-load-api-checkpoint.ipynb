```python
## input load data into the yaml file

import pandas as pd
import yaml

# Load time series load data from a CSV file
csv_file_path = './load data/Load data.csv'
load_data = pd.read_csv(csv_file_path, header=None)
desired_schedule = load_data[0].tolist()

# Load the original YAML file
yaml_file_path = './inputs/solar-wind-battery-api.yaml'
with open(yaml_file_path, 'r') as file:
    config = yaml.safe_load(file)

# Insert/rewrite the desired_schedule into the YAML configuration
config['site']['desired_schedule'] = desired_schedule

# Save the modified configuration back to the YAML file (or a new file if preferred)
modified_yaml_file_path = './inputs/solar-wind-battery-load-api.yaml'
with open(modified_yaml_file_path, 'w') as file:
    yaml.safe_dump(config, file, default_flow_style=False, sort_keys=False)
```

```python
# Create simulation, retrive the system parameters }from the .yaml file

hopp = HoppInterface("./inputs/solar-wind-battery-load-api.yaml")
```

```python
### this part is for mannually set up desired_schedule for load following dispatch, skip it if it is not needed. 

import numpy as np
import yaml

# Load the YAML configuration
hopp = load_yaml("./inputs/solar-wind-battery-load-api.yaml")

# Calculate the desired_schedule, and then input it into the site info (desired_schedule) parameter
baseload_limit_kw = float(30 * 1000)  # Baseline load in kW, convert from MW to kW
baseload_percent = 95.0  # percentage of minimum baseload requirement
DEFAULT_LOAD = baseload_limit_kw * np.ones((8760)) / 1000  # Generating an array for the whole year

hopp['site']['desired_schedule'] = DEFAULT_LOAD.tolist()  # Convert numpy array to list for YAML compatibility, after this, hopp will have 30 MW baseload for 8760 hours in the desire_schedule part
```

```python
# Run the Simulation. Simulate the hybrid renewable energy system for a specified number of years (in this case, 25 years).

hopp.simulate(project_life=25)
```

```python
# print simulation result

hybrid_plant = hopp.system

# collect results: energy output, npvs, installation costs etc.
annual_energies = hybrid_plant.annual_energies

npvs = hybrid_plant.net_present_values
cf = hybrid_plant.capacity_factors
revs = hybrid_plant.total_revenues

###
wind_installed_cost = hybrid_plant.wind.total_installed_cost
solar_installed_cost = hybrid_plant.pv.total_installed_cost
battery_installed_cost = hybrid_plant.battery.total_installed_cost
hybrid_installed_cost = hybrid_plant.grid.total_installed_cost

###
print("Wind Installed Cost: {}".format(wind_installed_cost))
print("Solar Installed Cost: {}".format(solar_installed_cost))
print("Battery Installed Cost: {}".format(battery_installed_cost))
print("Hybrid Installed Cost: {}".format(hybrid_installed_cost))

###
print("Wind NPV: {}".format(hybrid_plant.net_present_values.wind))
print("Solar NPV: {}".format(hybrid_plant.net_present_values.pv))
print("Battery NPV: {}".format(hybrid_plant.net_present_values.battery))
print("Hybrid NPV: {}".format(hybrid_plant.net_present_values.hybrid))

###
print("Annual Energies:")
print(annual_energies)

print("Net Present Values:")
print(npvs)

print("Total Revenues:")
print(revs)

###
print("Wind output after losses over gross output:",
      hybrid_plant.wind.value("annual_energy") / hybrid_plant.wind.value("annual_gross_energy"))
```
