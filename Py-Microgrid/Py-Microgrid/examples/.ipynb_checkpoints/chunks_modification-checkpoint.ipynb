```python
import pandas as pd
import os
import glob

# merge chunks into one dataframe
# Define the directory where the files are located
directory = "/home/z5142067/miniconda3/envs/microgrid/lib/python3.8/site-packages/hopp/examples/parallel_simulations/simulation_results_modularised/"

# Use glob to find all files that match the pattern 'lowest_cost_configs_chunk*.csv'
file_pattern = os.path.join(directory, "lowest_cost_configs_chunk*.csv")
file_list = glob.glob(file_pattern)

# Initialize an empty list to store DataFrames
dfs = []

# Loop through the list of files
for file_path in file_list:
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)
    
    # Append the DataFrame to the list
    dfs.append(df)

# Concatenate all DataFrames into one
combined_df = pd.concat(dfs, ignore_index=True)

# Define the output directory and ensure it exists
output_directory = os.path.join(directory, "combined_results_with_DR")
os.makedirs(output_directory, exist_ok=True)

# Save the combined DataFrame to a new CSV file in the combined_results directory
output_file_path = os.path.join(output_directory, "combined_lowest_cost_systems_heuristic.csv")
combined_df.to_csv(output_file_path, index=False)

print(f"Combined CSV saved to: {output_file_path}")
```

```python
## code to update yaml file load data.

import pandas as pd
import yaml
import os

# Step 1: Load the CSV file containing load data
load_data_path = './load data/Load data.csv'  # Adjust this path as per your setup
load_data_df = pd.read_csv(load_data_path)

# Step 2: Define the directory containing the YAML files
yaml_dir = './parallel_simulations/input_yaml'  # Adjust to your 'inputs' directory

# Step 3: Iterate through each YAML file in the directory
for yaml_file in os.listdir(yaml_dir):
    if yaml_file.endswith(".yaml"):
        yaml_file_path = os.path.join(yaml_dir, yaml_file)
        
        # Step 4: Load the YAML file
        with open(yaml_file_path, 'r') as file:
            yaml_content = yaml.safe_load(file)
            if yaml_content is None:  # If the file is empty or invalid
                yaml_content = {}  # Initialize as an empty dictionary if loading fails
        
        # Step 5: Update the 'desired_schedule' field with all rows from the CSV
        # Assuming the 'desired_schedule' corresponds to the entire column in the CSV (8760 values)
        # You can specify the column name if necessary, otherwise it takes the first column by default
        load_values = load_data_df.iloc[:, 0].tolist()  # Assuming the first column contains the 8760 values
        
        # Navigate to 'site' > 'desired_schedule' and update the values
        if 'site' in yaml_content and 'desired_schedule' in yaml_content['site']:
            yaml_content['site']['desired_schedule'] = load_values
        else:
            print(f"'desired_schedule' not found in {yaml_file}. Skipping.")
        
        # Step 6: Write the updated content back to the YAML file
        with open(yaml_file_path, 'w') as file:
            yaml.dump(yaml_content, file)
```

```python
# rewrite yaml files to change the battery dispatch option

import yaml
import os

# Directory where the YAML files are located
yaml_dir = './parallel_simulations/input_yaml'

# Function to update battery_dispatch in the YAML file
def update_battery_dispatch(yaml_file):
    with open(yaml_file, 'r') as file:
        config = yaml.safe_load(file)

    # Update the battery_dispatch field
    config['config']['dispatch_options']['battery_dispatch'] = 'predictive_demand_response'

    # Write the updated configuration back to the file
    with open(yaml_file, 'w') as file:
        yaml.safe_dump(config, file)

# Iterate through the YAML directory and update all files
for filename in os.listdir(yaml_dir):
    if filename.endswith('.yaml'):
        yaml_file = os.path.join(yaml_dir, filename)
        update_battery_dispatch(yaml_file)

print("Battery dispatch updated in all YAML files.")
```

```python
# create more yaml file chunks for simulation

import yaml
import os
import shutil

# Directory where the original YAML files are stored
yaml_dir = '../parallel_simulations/input_yaml'

# Function to generate new YAML files with updated 'name'
def generate_yaml_files(start, count, output_dir):
    for i in range(start, start + count):
        # Determine the file name
        new_file_name = f'input_file_chunk_{i}.yaml'
        new_file_path = os.path.join(output_dir, new_file_name)

        # Copy an existing file (you can pick one as a template)
        existing_file_path = os.path.join(output_dir, 'input_file_chunk_0.yaml')  # Adjust as needed

        # Load the YAML data from the existing file
        with open(existing_file_path, 'r') as file:
            config = yaml.safe_load(file)

        # Update the 'name' field
        config['name'] = f'input_file_chunk_{i}'

        # Write the updated data to the new YAML file
        with open(new_file_path, 'w') as file:
            yaml.safe_dump(config, file)

    print(f'{count} YAML files generated starting from input_file_chunk_{start}.')

# Specify the starting number and how many files you want to create
start_number = 0  # Change this to the starting file number
number_of_files = 45  # Change this to the number of files to create

# Create the YAML files
generate_yaml_files(start_number, number_of_files, yaml_dir)
```

```python
# create more notebook chunks for simulation 

import nbformat
import os

def create_chunk_notebooks(input_notebook_path, start_chunk, end_chunk, output_directory):
    # Load the input Jupyter notebook
    with open(input_notebook_path, 'r', encoding='utf-8') as f:
        notebook = nbformat.read(f, as_version=4)

    # Function to replace the old chunk number with the new one in the entire notebook content
    def replace_chunk_number(content, old_number, new_number):
        return content.replace(f'chunk_{old_number}', f'chunk_{new_number}')

    # Extract the current chunk number from the input notebook name or content
    current_chunk_number = int(input_notebook_path.split('_')[-1].split('.')[0])

    # Iterate over the range of chunk numbers to create new notebooks
    for chunk_number in range(start_chunk, end_chunk + 1):
        new_notebook = notebook.copy()
        
        # Replace the old chunk number (from the previous chunk) with the current new chunk number
        for cell in new_notebook['cells']:
            if cell['cell_type'] == 'code' or cell['cell_type'] == 'markdown':
                cell['source'] = replace_chunk_number(cell['source'], current_chunk_number, chunk_number)
        
        # Save the new notebook with the updated chunk number in the file name
        output_notebook_path = os.path.join(output_directory, f'simulation_chunk_{chunk_number}.ipynb')
        with open(output_notebook_path, 'w', encoding='utf-8') as f:
            nbformat.write(new_notebook, f)

        print(f'Created notebook: {output_notebook_path}')
        
        # Update the current_chunk_number to the new one for next iteration
        current_chunk_number = chunk_number

    print(f"All notebooks from chunk {start_chunk} to {end_chunk} created successfully.")

# Example usage
input_notebook_path = './parallel_simulations/scripts_modularised/simulation_chunk_36.ipynb'  # Path to the input notebook (change as needed)
start_chunk = 36 # The chunk number to start with
end_chunk = 44   # The chunk number to end with (change as needed)
output_directory = './parallel_simulations/scripts_modularised'  # Directory to save the new notebooks

create_chunk_notebooks(input_notebook_path, start_chunk, end_chunk, output_directory)
```

```python
import nbformat
import os
import re

def update_notebooks(directory):
    # Iterate through all files in the directory
    for filename in os.listdir(directory):
        if filename.endswith('.ipynb') and filename.startswith('simulation_chunk_'):
            file_path = os.path.join(directory, filename)
            
            # Extract chunk number from filename
            chunk_number = re.search(r'simulation_chunk_(\d+)', filename).group(1)
            
            # Load the notebook
            with open(file_path, 'r', encoding='utf-8') as f:
                notebook = nbformat.read(f, as_version=4)
            
            # Flag to check if any changes were made
            changes_made = False
            
            # Iterate through cells and update the specific line
            for cell in notebook['cells']:
                if cell['cell_type'] == 'code':
                    # Update the pattern to match the structure with '_simple.csv'
                    updated_source = re.sub(
                        r'lowest_cost_configs_chunk_\d+_heuristic\.csv',
                        f'lowest_cost_configs_chunk_{chunk_number}_PDR.csv',
                        cell['source']
                    )
                    if updated_source != cell['source']:
                        cell['source'] = updated_source
                        changes_made = True
            
            # If changes were made, save the notebook
            if changes_made:
                with open(file_path, 'w', encoding='utf-8') as f:
                    nbformat.write(notebook, f)
                print(f"Updated: {filename}")
            else:
                print(f"No changes needed: {filename}")

# Directory containing the notebooks
directory = './parallel_simulations/scripts_modularised'

# Run the update function
update_notebooks(directory)
print("All notebooks have been processed.")
```
