```python
"""
HOPP batch simulation script. It is used to compare LCOE with HOMER results (benchmarking).
It uses the exact system config from HOMER results to simulate hybrid microgrid using HOPP.
Handles both individual simulations and batch processing with exact LCOE calculations.
"""

import os
import numpy as np
import pandas as pd
from pathlib import Path
import yaml
import logging
from tqdm import tqdm
from typing import Dict, Any

from hopp.simulation import HoppInterface
from hopp.utilities import ConfigManager
from hopp.tools.optimization import LoadAnalyzer
from hopp.tools.analysis.bos import EconomicCalculator
from hopp.utilities.keys import set_developer_nrel_gov_key

# Configure logging
logging.basicConfig(
    filename='simulation.log',
    filemode='a',
    format='%(asctime)s - %(levelname)s - %(message)s',
    level=logging.INFO
)

def calculate_present_value(future_value: float, discount_rate: float, project_lifetime: int) -> float:
    """Calculate present value of future costs."""
    return future_value / ((1 + discount_rate) ** project_lifetime)

def calculate_lcoe(total_system_cost: float, total_system_generation_over_lifetime: float, 
                  discount_rate: float, project_lifetime: int) -> float:
    """Calculate Levelized Cost of Energy."""
    present_value_total_system_costs = calculate_present_value(
        total_system_cost, discount_rate, project_lifetime
    )
    present_value_total_system_generation = calculate_present_value(
        total_system_generation_over_lifetime, discount_rate, project_lifetime
    )
    return present_value_total_system_costs / present_value_total_system_generation

def run_simulation(config: Dict, 
                  yaml_file_path: str,
                  enable_flexible_load: bool = True,
                  max_load_reduction_percentage: float = 0.2,
                  project_lifetime: int = 25,
                  discount_rate: float = 0.0588) -> Dict[str, Any]:
    """
    Run single simulation with given configuration.
    """
    try:
        # Initialize components
        config_manager = ConfigManager()
        economic_calculator = EconomicCalculator(discount_rate, project_lifetime)
        load_analyzer = LoadAnalyzer(
            max_load_reduction_percentage=max_load_reduction_percentage,
            enable_flexible_load=enable_flexible_load
        )

        # Save temporary configuration
        temp_yaml_path = 'temp_config.yaml'
        with open(temp_yaml_path, 'w') as file:
            yaml.dump(config, file)

        # Run simulation
        hopp = HoppInterface(temp_yaml_path)
        hopp.simulate(project_life=project_lifetime)
        hybrid_plant = hopp.system

        # Calculate generation values
        pv_total_generation = np.sum(hybrid_plant.generation_profile.pv)
        wind_total_generation = np.sum(hybrid_plant.generation_profile.wind)
        battery_total_generation = np.sum(hybrid_plant.generation_profile.battery)
        genset_total_generation = np.sum(hybrid_plant.generation_profile.grid)
        total_system_generation = pv_total_generation + wind_total_generation + genset_total_generation

        # Calculate genset metrics
        genset_capacity_kw = hybrid_plant.grid.interconnect_kw
        genset_operational_hours_per_year = np.sum(np.array(hybrid_plant.grid.generation_profile) > 0) / project_lifetime
        generator_operational_life_hours = 15000
        generator_operational_life_years = generator_operational_life_hours / genset_operational_hours_per_year
        num_genset_replacements = float(project_lifetime / generator_operational_life_years) - 1

        # Calculate component costs
        # Genset costs
        genset_install_cost = genset_capacity_kw * 500
        genset_replacement_cost = num_genset_replacements * genset_capacity_kw * 500
        genset_om_cost = 0.03 * genset_capacity_kw * genset_operational_hours_per_year * project_lifetime
        fuel_consumption = genset_total_generation * 0.250
        fuel_cost = fuel_consumption * 1.20
        co2_emissions = fuel_consumption * 2.618
        total_genset_cost = genset_install_cost + genset_replacement_cost + genset_om_cost + fuel_cost

        # Battery costs
        battery_capacity_kwh = config['technologies']['battery']['system_capacity_kwh']
        battery_installed_cost = hybrid_plant.battery.total_installed_cost
        battery_replacement_cost = ((project_lifetime/15) - 1) * battery_capacity_kwh * (battery_installed_cost/battery_capacity_kwh)
        battery_om_cost = 10 * battery_capacity_kwh * project_lifetime
        total_battery_cost = battery_installed_cost + battery_replacement_cost + battery_om_cost

        # PV and Wind costs
        pv_installed_cost = hybrid_plant.pv.total_installed_cost
        pv_om_cost = 10 * hybrid_plant.system_capacity_kw.pv * project_lifetime
        total_pv_cost = pv_installed_cost + pv_om_cost

        wind_installed_cost = hybrid_plant.wind.total_installed_cost
        wind_om_cost = 40 * hybrid_plant.system_capacity_kw.wind * project_lifetime
        total_wind_cost = wind_installed_cost + wind_om_cost

        # Total system cost
        total_system_cost = total_genset_cost + total_battery_cost + total_pv_cost + total_wind_cost

        # Create DataFrame for load analysis
        df = pd.DataFrame({
            'PV Generation (kW)': np.array(hybrid_plant.generation_profile.pv[:8760]),
            'Wind Generation (kW)': np.array(hybrid_plant.generation_profile.wind[:8760]),
            'Genset Generation (kW)': np.array(hybrid_plant.generation_profile.grid[:8760]),
            'Original Battery Generation (kW)': np.array(hybrid_plant.generation_profile.battery[:8760]),
            'Original Load (kW)': np.array(hybrid_plant.site.desired_schedule[:8760]) * 1000
        })

        # Calculate metrics
        metrics = load_analyzer.calculate_performance_metrics(df, project_lifetime)

        # Calculate LCOE - using your exact method
        lcoe = calculate_lcoe(total_system_cost, metrics['Total Load Served (kWh)'],
                            discount_rate, project_lifetime)

        return {
            "Latitude": config['site']['data']['lat'],    
            "Longitude": config['site']['data']['lon'],   
            "PV Capacity (kW)": config['technologies']['pv']['system_capacity_kw'],
            "Wind Turbine Capacity (kW)": config['technologies']['wind']['num_turbines'] * 1000,
            "Genset Capacity (kW)": config['technologies']['grid']['interconnect_kw'],
            "Battery Energy Capacity (kWh)": config['technologies']['battery']['system_capacity_kwh'],
            "Battery Power Capacity (kW)": config['technologies']['battery']['system_capacity_kw'],
            "Total System Generation (kWh)": total_system_generation,
            "Total PV Generation (kWh)": pv_total_generation,
            "Total Wind Generation (kWh)": wind_total_generation,
            "Total Genset Generation (kWh)": genset_total_generation,
            "Total Battery Generation (kWh)": battery_total_generation,
            "Total Load Served (kWh)": metrics['Total Load Served (kWh)'],
            "Total CO2 emissions (tonne)": co2_emissions/1000,
            "System Cost ($)": total_system_cost,
            "System NPC ($)": calculate_present_value(total_system_cost, discount_rate, project_lifetime),
            "System LCOE ($/kWh)": lcoe,
            **metrics
        }

    except Exception as e:
        logging.error(f"Simulation failed: {str(e)}")
        return None
    
    finally:
        if os.path.exists(temp_yaml_path):
            os.remove(temp_yaml_path)

def process_locations(input_data: pd.DataFrame, yaml_file_path: str,
                     enable_flexible_load: bool = True,
                     max_load_reduction_percentage: float = 0.2) -> list:
    """
    Process multiple locations from input data.
    """
    results = []
    failed_locations = []
    
    # Load base configuration
    config_manager = ConfigManager()
    base_config = config_manager.load_yaml_safely(yaml_file_path)
    
    print(f"\nProcessing {len(input_data)} locations...")
    
    with tqdm(total=len(input_data), desc="Processing Locations") as pbar:
        for index, row in input_data.iterrows():
            try:
                # Update configuration for current location
                config = base_config.copy()
                
                # Update site information
                config['site']['data']['lat'] = row['DEPOSIT_LATITUDE_WGS84']
                config['site']['data']['lon'] = row['DEPOSIT_LONGITUDE_WGS84']
                
                # Update resource file paths
                latitude = row['DEPOSIT_LATITUDE_WGS84']
                longitude = row['DEPOSIT_LONGITUDE_WGS84']
                
                solar_filename = f"{latitude}_{longitude}_psmv3_60_2020.csv"
                wind_filename = f"{latitude}_{longitude}_NASA_2020_60min_50m.srw"
                
                solar_path = config['site'].get('solar_resource_file')
                wind_path = config['site'].get('wind_resource_file')
                
                if solar_path:
                    solar_dir = Path(solar_path).parent
                    config['site']['solar_resource_file'] = str(solar_dir / solar_filename)
                
                if wind_path:
                    wind_dir = Path(wind_path).parent
                    config['site']['wind_resource_file'] = str(wind_dir / wind_filename)
                
                # Update system capacities
                config['technologies']['pv']['system_capacity_kw'] = float(row['Architecture/PV (kW)'])
                config['technologies']['wind']['num_turbines'] = int(row['Architecture/Wind 1MW'])
                config['technologies']['battery']['system_capacity_kwh'] = float(row['Architecture/Battery 1MWh (#)']) * 1000
                config['technologies']['grid']['interconnect_kw'] = float(row['Architecture/Gen (kW)'])
                
                # Run simulation
                result = run_simulation(
                    config=config,
                    yaml_file_path=yaml_file_path,
                    enable_flexible_load=enable_flexible_load,
                    max_load_reduction_percentage=max_load_reduction_percentage
                )
                
                if result:
                    result.update({
                        'Latitude': row['DEPOSIT_LATITUDE_WGS84'],
                        'Longitude': row['DEPOSIT_LONGITUDE_WGS84'],
                        'Deposit_ID': row.get('DEPOSIT_UID', f'DEP_{index}'),
                        'Configuration_Index': index
                    })
                    results.append(result)
                else:
                    failed_locations.append((index, "Simulation failed"))
                    
            except Exception as e:
                failed_locations.append((index, str(e)))
                logging.error(f"Error processing location {index}: {str(e)}")
            
            finally:
                pbar.update(1)
    
    # Log results
    print(f"\nProcessed {len(results)} locations successfully")
    if failed_locations:
        print(f"Failed to process {len(failed_locations)} locations")
        pd.DataFrame(failed_locations, columns=['Index', 'Reason']).to_csv(
            'failed_locations.csv', index=False
        )
    
    return results

def main():
    """Main execution function."""
    try:
        # Configuration
        input_csv_path = 'FL hybrid MGs + LF collection_v2.csv'
        yaml_file_path = './examples_old/inputs/test file 2-api alternative.yaml'
        output_csv_path = 'simulation_results_HOPP_vs_HOMER_LF.csv'
        
        # Set API key
        set_developer_nrel_gov_key('3F5fBErsd9gKIPEdQpkWHNIhNj7gZG3j0y3lkzew')
        
        # Load and validate input data
        print("Loading input data...")
        input_data = pd.read_csv(input_csv_path)
        print(f"Total locations in input file: {len(input_data)}")
        
        # Validate required columns
        required_columns = [
            'DEPOSIT_LATITUDE_WGS84',
            'DEPOSIT_LONGITUDE_WGS84',
            'Architecture/PV (kW)',
            'Architecture/Wind 1MW',
            'Architecture/Battery 1MWh (#)',
            'Architecture/Gen (kW)'
        ]
        
        # Check for missing columns
        missing_columns = [col for col in required_columns if col not in input_data.columns]
        if missing_columns:
            raise ValueError(f"Missing required columns: {missing_columns}")
        
        # Clean input data
        initial_count = len(input_data)
        input_data = input_data.dropna(subset=required_columns)
        if len(input_data) < initial_count:
            print(f"Removed {initial_count - len(input_data)} rows with missing values")
            print(f"Proceeding with {len(input_data)} valid locations")
        
        # Process all locations
        print("\nStarting simulations...")
        results = process_locations(
            input_data=input_data,
            yaml_file_path=yaml_file_path,
            enable_flexible_load=False,  # disable flexible load for precise LCOE comparison with HOMER
            max_load_reduction_percentage=0.2
        )
        
        # Save results
        if results:
            results_df = pd.DataFrame(results)
            results_df.to_csv(output_csv_path, index=False)
            
            # Print summary statistics
            print(f"\nResults saved to: {output_csv_path}")
            print("\nSummary statistics:")
            print(f"Total configurations processed: {len(results)}")
            print(f"Average LCOE: ${results_df['System LCOE ($/kWh)'].mean():.4f}/kWh")
            print(f"Min LCOE: ${results_df['System LCOE ($/kWh)'].min():.4f}/kWh")
            print(f"Max LCOE: ${results_df['System LCOE ($/kWh)'].max():.4f}/kWh")
        else:
            print("No results to save.")
    
    except Exception as e:
        logging.exception("Error in main execution")
        print(f"\nError: {str(e)}")
        print("Check 'simulation.log' for details.")

if __name__ == "__main__":
    main()
```

```python
"""
Modularised test script for running a direct simulation from YAML without optimization.
Allows testing with and without flexible load management.
"""

import os
import numpy as np
import pandas as pd
from typing import Dict, Any

from hopp.simulation import HoppInterface
from hopp.utilities import ConfigManager
from hopp.tools.optimization import LoadAnalyzer
from hopp.tools.analysis.bos import EconomicCalculator

def run_direct_simulation(yaml_file_path: str,
                         enable_flexible_load: bool = True,
                         max_load_reduction_percentage: float = 0.2,
                         project_lifetime: int = 25,
                         discount_rate: float = 0.0588) -> Dict[str, Any]:
    """
    Run direct simulation with given YAML configuration.
    
    Args:
        yaml_file_path: Path to YAML configuration file
        enable_flexible_load: Whether to enable flexible load management
        max_load_reduction_percentage: Maximum percentage for load reduction
        project_lifetime: Project lifetime in years
        discount_rate: Discount rate for economic calculations
        
    Returns:
        Dictionary containing simulation results
    """
    try:
        # Initialize components
        config_manager = ConfigManager()
        economic_calculator = EconomicCalculator(discount_rate, project_lifetime)
        load_analyzer = LoadAnalyzer(
            max_load_reduction_percentage=max_load_reduction_percentage,
            enable_flexible_load=enable_flexible_load
        )
        
        # Load configuration
        config = config_manager.load_yaml_safely(yaml_file_path)
        print(f"\nRunning simulation with configuration from: {yaml_file_path}")
        print(f"Flexible Load: {'Enabled' if enable_flexible_load else 'Disabled'}")
        if enable_flexible_load:
            print(f"Maximum Load Reduction: {max_load_reduction_percentage*100}%")
        
        # Run simulation
        hopp = HoppInterface(yaml_file_path)
        hopp.simulate(project_life=project_lifetime)
        
        hybrid_plant = hopp.system
        
        # Calculate generation values
        pv_total_generation = np.sum(hybrid_plant.generation_profile.pv)
        wind_total_generation = np.sum(hybrid_plant.generation_profile.wind)
        battery_total_generation = np.sum(hybrid_plant.generation_profile.battery)
        genset_total_generation = np.sum(hybrid_plant.generation_profile.grid)
        total_system_generation = pv_total_generation + wind_total_generation + genset_total_generation

        # Calculate genset metrics
        genset_capacity_kw = hybrid_plant.grid.interconnect_kw
        genset_operational_hours_per_year = np.sum(np.array(hybrid_plant.grid.generation_profile) > 0) / project_lifetime
        generator_operational_life_hours = 15000
        generator_operational_life_years = generator_operational_life_hours / genset_operational_hours_per_year
        num_genset_replacements = float(project_lifetime / generator_operational_life_years) - 1

        # Calculate component costs
        # Genset costs
        genset_install_cost = genset_capacity_kw * 500
        genset_replacement_cost = num_genset_replacements * genset_capacity_kw * 500
        genset_om_cost = 0.03 * genset_capacity_kw * genset_operational_hours_per_year * project_lifetime
        fuel_consumption = genset_total_generation * 0.250
        fuel_cost = fuel_consumption * 1.20
        co2_emissions = fuel_consumption * 2.618
        total_genset_cost = genset_install_cost + genset_replacement_cost + genset_om_cost + fuel_cost
        
        # Battery costs
        battery_capacity_kwh = config['technologies']['battery']['system_capacity_kwh']
        battery_installed_cost = hybrid_plant.battery.total_installed_cost
        battery_replacement_cost = ((project_lifetime/15) - 1) * battery_capacity_kwh * (battery_installed_cost/battery_capacity_kwh)
        battery_om_cost = 10 * battery_capacity_kwh * project_lifetime
        total_battery_cost = battery_installed_cost + battery_replacement_cost + battery_om_cost
        
        # PV and Wind costs
        pv_installed_cost = hybrid_plant.pv.total_installed_cost
        pv_om_cost = 10 * hybrid_plant.system_capacity_kw.pv * project_lifetime
        total_pv_cost = pv_installed_cost + pv_om_cost
        
        wind_installed_cost = hybrid_plant.wind.total_installed_cost
        wind_om_cost = 40 * hybrid_plant.system_capacity_kw.wind * project_lifetime
        total_wind_cost = wind_installed_cost + wind_om_cost

        # Total system cost
        total_system_cost = total_genset_cost + total_battery_cost + total_pv_cost + total_wind_cost

        # Create DataFrame for load analysis
        df = pd.DataFrame({
            'PV Generation (kW)': np.array(hybrid_plant.generation_profile.pv[:8760]),
            'Wind Generation (kW)': np.array(hybrid_plant.generation_profile.wind[:8760]),
            'Genset Generation (kW)': np.array(hybrid_plant.generation_profile.grid[:8760]),
            'Original Battery Generation (kW)': np.array(hybrid_plant.generation_profile.battery[:8760]),
            'Original Load (kW)': np.array(hybrid_plant.site.desired_schedule[:8760]) * 1000
        })

        # Calculate metrics based on flexible load setting
        metrics = load_analyzer.calculate_performance_metrics(df, project_lifetime)
        
        # Calculate LCOE
        lcoe = economic_calculator.calculate_lcoe(total_system_cost, metrics['Total Load Served (kWh)'])

        return {
            "PV Capacity (kW)": config['technologies']['pv']['system_capacity_kw'],
            "Wind Turbine Capacity (kW)": config['technologies']['wind']['num_turbines'] * 1000,
            "Genset Capacity (kW)": config['technologies']['grid']['interconnect_kw'],
            "Battery Energy Capacity (kWh)": config['technologies']['battery']['system_capacity_kwh'],
            "Battery Power Capacity (kW)": config['technologies']['battery']['system_capacity_kw'],
            "Total System Generation (kWh)": total_system_generation,
            "Total PV Generation (kWh)": pv_total_generation,
            "Total Wind Generation (kWh)": wind_total_generation,
            "Total Genset Generation (kWh)": genset_total_generation,
            "Total Battery Generation (kWh)": battery_total_generation,
            "Total CO2 emissions (tonne)": co2_emissions/1000,
            "System Cost ($)": total_system_cost,
            "System NPC ($)": economic_calculator.calculate_present_value(total_system_cost),
            "System LCOE ($/kWh)": lcoe,
            **metrics
        }

    except Exception as e:
        print(f"Simulation failed: {str(e)}")
        return {}

if __name__ == "__main__":
    # Example usage
    yaml_file = "./examples_old/inputs/test file 2-api alternative.yaml"
    
    # Run without flexible load
    print("\nRunning simulation without flexible load...")
    results_no_flex = run_direct_simulation(yaml_file, enable_flexible_load=False)
    
    # Run with flexible load
    #print("\nRunning simulation with flexible load...")
    #results_flex = run_direct_simulation(yaml_file, enable_flexible_load=True)
```

```python
'''
Plot HOPP hybrid microgrids simulation output for 1 week, Load Following dispatch
Deposit: AUS.NSW.333485 (lat: -31.1629, lon: 145.6538)
'''

import matplotlib.pyplot as plt
from hopp.simulation.hybrid_simulation import HybridSimulation
import numpy as np

def plot_hybrid_system(hybrid: HybridSimulation,
                      start_day: int = 0,
                      n_days: int = 7,
                      plot_filename: str = None,
                      font_size: int = 14,
                      power_scale: float = 1/1000,
                      solar_color='tab:orange',
                      wind_color='tab:green',
                      discharge_color='tab:blue',
                      charge_color='tab:red',
                      gen_color='tab:brown'
                      ):
    
    if not hasattr(hybrid, 'dispatch_builder'):
        raise AttributeError("Simulation with dispatch must be called before plotting generation profile.")
        
    start = start_day * hybrid.site.n_periods_per_day
    end = start + n_days * hybrid.site.n_periods_per_day
    time = range(start, end)
    time_slice = slice(start, end)
    
    fig = plt.figure(figsize=(15, 10), dpi=300)
    
    # First subplot (a) - Generation profiles
    ax1 = plt.subplot(2, 1, 1)
    plt.tick_params(which='both', labelsize=font_size)
    
    # Plot generation sources
    if hybrid.pv:
        solar = [x * power_scale for x in hybrid.pv.generation_profile[time_slice]]
        plt.plot(time, solar, color=solar_color, label='PV Generation')
    
    if hybrid.wind:
        wind = [x * power_scale for x in hybrid.wind.generation_profile[time_slice]]
        plt.plot(time, wind, color=wind_color, label='Wind Generation')
    
    genset_output = [p * power_scale for p in hybrid.grid.generation_profile[time_slice]]
    plt.plot(time, genset_output, color=gen_color, label='Genset Generation')
    
    plt.xlim([start, end])
    plt.ylim(bottom=0)  # Set the bottom limit to 0
    ax1.xaxis.set_ticks(list(range(start, end, hybrid.site.n_periods_per_day)))
    
    # Format y-axis ticks as integers for subplot (a)
    ax1.set_yticks(range(0, 21, 5))  # 0 to 20 in steps of 5
    ax1.yaxis.set_major_formatter(plt.FormatStrFormatter('%d'))
    
    plt.grid(False)
    plt.ylabel('Power (MW)', fontsize=font_size)
    plt.legend(fontsize=font_size-2, loc='center left', bbox_to_anchor=(1, 0.5))
    plt.text(-0.1, 1.05, 'a)', transform=ax1.transAxes, fontsize=font_size, fontweight='bold')
    
    # Second subplot (b) - Battery power flow, SOC and Desired Load
    ax2 = plt.subplot(2, 1, 2)
    plt.tick_params(which='both', labelsize=font_size)
    
    # Calculate discharge and charge
    discharge = [(p > 0) * p * power_scale for p in hybrid.battery.outputs.P[time_slice]]
    charge = [(p < 0) * p * power_scale for p in hybrid.battery.outputs.P[time_slice]]
    
    # Create first axis and its plots
    ax2.bar(time, discharge, width=0.9, color=discharge_color, 
            edgecolor='white', label='Battery Discharge', alpha=0.7)
    ax2.bar(time, charge, width=0.9, color=charge_color, 
            edgecolor='white', label='Battery Charge', alpha=0.7)
    
    # Plot desired load if available
    if hybrid.site.follow_desired_schedule:
        desired_load = [p for p in hybrid.site.desired_schedule[time_slice]]
        ax2.plot(time, desired_load, 'b--', label='Desired Load')
    
    ax2.set_xlim([start, end])
    ax2.xaxis.set_ticks(list(range(start, end, hybrid.site.n_periods_per_day)))
    
    # Make y-axis symmetric for subplot (b) with fixed limits at ±15
    ax2.set_ylim(-10, 15)
    ax2.set_yticks(np.arange(-10, 16, 5))  # -15 to +15 in steps of 5
    
    ax2.grid(False)
    ax2.axhline(y=0, color='black', linestyle='-', alpha=0.3)
    ax2.set_ylabel('Power (MW)', fontsize=font_size)
    
    # Create second axis for SOC
    ax3 = ax2.twinx()
    ax3.plot(time, hybrid.battery.outputs.SOC[time_slice], 'k', label='State-of-Charge')
    ax3.set_ylabel('State-of-Charge (%)', fontsize=font_size)
    
    # Create two rows of legends next to the right y-axis
    leg1 = ax2.legend(fontsize=font_size-2, loc='center left', 
                     bbox_to_anchor=(1.05, 0.6))
    leg2 = ax3.legend(fontsize=font_size-2, loc='center left', 
                     bbox_to_anchor=(1.05, 0.4))
    
    plt.text(-0.1, 1.05, 'b)', transform=ax2.transAxes, fontsize=font_size, fontweight='bold')
    plt.xlabel('Time (hours)', fontsize=font_size)
    
    # Adjust the subplot parameters to make room for the legend
    plt.subplots_adjust(right=0.85, hspace=0.3)
    
    # Save and show plot
    if plot_filename:
        plt.savefig(plot_filename, bbox_inches='tight', dpi=300)
    else:
        plt.savefig('HOPP_hybrid_microgrid.png', bbox_inches='tight', dpi=300)
    
    plt.show()
```

```python
import pandas as pd
import numpy as np
import os

# Assuming the generation profiles are lists and converting them to DataFrames
pv_list = hybrid_plant.generation_profile.pv[:8760]
wind_list = hybrid_plant.generation_profile.wind[:8760]
genset_list = hybrid_plant.generation_profile.grid[:8760]
battery_list = hybrid_plant.generation_profile.battery[:8760]
load_list = hybrid_plant.site.desired_schedule

# Convert the lists to DataFrames
pv_df = pd.DataFrame(pv_list, columns=['PV Generation (kW)'])
wind_df = pd.DataFrame(wind_list, columns=['Wind Generation (kW)'])
genset_df = pd.DataFrame(genset_list, columns=['Genset Generation (kW)'])
battery_df = pd.DataFrame(battery_list, columns=['Original Battery Generation (kW)'])

# Convert the load list from MW to kW by multiplying by 1000
load_list_kW = [load * 1000 for load in load_list]
load_df = pd.DataFrame(load_list_kW, columns=['Original Load (kW)'])

# Combine these DataFrames into a single DataFrame
combined_df = pd.concat([pv_df, wind_df, genset_df, battery_df, load_df], axis=1)

# Define the function to apply flexible load and adjust battery
def apply_flexible_load_and_adjust_battery(row):
    pv = row['PV Generation (kW)']
    wind = row['Wind Generation (kW)']
    genset = row['Genset Generation (kW)']
    battery = row['Original Battery Generation (kW)']
    original_load = row['Original Load (kW)']
    
    total_generation_without_battery = pv + wind + genset
    
    # First, adjust battery if it's causing a deficit
    if battery < 0:  # Battery is charging
        if total_generation_without_battery + battery < original_load:  # Charging causes deficit
            battery = 0  # Prevent charging
    
    adjusted_generation = total_generation_without_battery + battery
    
    # Define the maximum load reduction percentage
    load_reduction_percentage = 0.20  # 20% deferrable load
    
    # Now apply flexible load if there's still a deficit
    if adjusted_generation < original_load:
        deficit = original_load - adjusted_generation
        max_load_reduction = load_reduction_percentage * original_load  # Maximum reduction based on the defined percentage
        load_reduction = min(deficit, max_load_reduction)
        adjusted_load = original_load - load_reduction
    else:
        adjusted_load = original_load
    
    new_deficit = adjusted_generation - adjusted_load
    
    return pd.Series({
        'Adjusted Battery Generation (kW)': battery,
        'Adjusted Load (kW)': adjusted_load,
        'Adjusted Deficit (kW)': new_deficit
    })

# Apply the flexible load and battery adjustment
adjusted_results = combined_df.apply(apply_flexible_load_and_adjust_battery, axis=1)

# Add the new columns to the original DataFrame
combined_df['Adjusted Battery Generation (kW)'] = adjusted_results['Adjusted Battery Generation (kW)']
combined_df['Adjusted Load (kW)'] = adjusted_results['Adjusted Load (kW)']
combined_df['Adjusted Deficit (kW)'] = adjusted_results['Adjusted Deficit (kW)']

# Recalculate total generation with adjusted battery output
combined_df['Adjusted Total Generation (kW)'] = combined_df['PV Generation (kW)'] + \
                                                combined_df['Wind Generation (kW)'] + \
                                                combined_df['Genset Generation (kW)'] + \
                                                combined_df['Adjusted Battery Generation (kW)']

# Calculate the initial deficit (before adjustments)
combined_df['Initial Deficit (kW)'] = combined_df['PV Generation (kW)'] + \
                                      combined_df['Wind Generation (kW)'] + \
                                      combined_df['Genset Generation (kW)'] + \
                                      combined_df['Original Battery Generation (kW)'] - \
                                      combined_df['Original Load (kW)']

# Identify the original rows with deficits
original_deficits = combined_df[combined_df['Initial Deficit (kW)'] < 0]

# Check for remaining negative deficits after adjustment
remaining_negative_deficits = combined_df[combined_df['Adjusted Deficit (kW)'] < 0]

# Define the output directory and ensure it exists
output_dir = "./test_results"
os.makedirs(output_dir, exist_ok=True)

# Write the combined DataFrame to a CSV file
csv_file_path = os.path.join(output_dir, "gen_load_compariso_PDR.csv")
combined_df.to_csv(csv_file_path, index=False)

# Provide the path to the user
print(f"Data saved to: {csv_file_path}")

# Report the number and rows of original negative deficit values
num_original_deficits = len(original_deficits)
original_deficit_rows = original_deficits.index.tolist()
original_deficit_rows_adjusted = [row + 2 for row in original_deficit_rows]

# Report the number and rows of remaining negative deficit values
num_remaining_negative_deficits = len(remaining_negative_deficits)
remaining_negative_deficit_rows = remaining_negative_deficits.index.tolist()
remaining_negative_deficit_rows_adjusted = [row + 2 for row in remaining_negative_deficit_rows]

# Calculate the number of deficit hours fixed by the adjustments
hours_fixed = num_original_deficits - num_remaining_negative_deficits

# Print the results
print(f"Original deficits found in {num_original_deficits} rows.")
print(f"Rows with original deficits: {original_deficit_rows_adjusted}")

if num_remaining_negative_deficits > 0:
    print(f"Remaining negative deficits found in {num_remaining_negative_deficits} rows after adjustments.")
    print(f"Rows with remaining negative deficits: {remaining_negative_deficit_rows_adjusted}")
else:
    print("No remaining negative deficits found after adjustments.")

print(f"Number of deficit hours fixed by adjustments: {hours_fixed}")

# Calculate total load reduction
total_load_reduction = (combined_df['Original Load (kW)'] - combined_df['Adjusted Load (kW)']).sum()
total_load_reduction_percentage = (total_load_reduction / combined_df['Original Load (kW)'].sum()) * 100

print(f"Total load reduction: {total_load_reduction:.2f} kWh")
print(f"Total load reduction percentage: {total_load_reduction_percentage:.2f}%")

# Calculate battery charging prevention
total_charging_prevented = (combined_df['Original Battery Generation (kW)'] - combined_df['Adjusted Battery Generation (kW)']).sum()
print(f"Total battery charging prevented: {total_charging_prevented:.2f} kWh")

# Calculate energy not served
total_energy_not_served = combined_df['Adjusted Deficit (kW)'][combined_df['Adjusted Deficit (kW)'] < 0].sum() * -1
print(f"Total energy not served after adjustments: {total_energy_not_served:.2f} kWh")

# Calculate the percentage of demand met
total_original_demand = combined_df['Original Load (kW)'].sum()
total_adjusted_demand = combined_df['Adjusted Load (kW)'].sum()
demand_met_percentage = (total_adjusted_demand - total_energy_not_served) / total_adjusted_demand * 100
print(f"Percentage of adjusted demand met: {demand_met_percentage:.2f}%")
```

```python
## LCOE = present value of total system cost/present value of total load served
## original version, where installed cost of each component is coming from HOPP, e.g. hybrid_plant.pv.total_installed_cost
import numpy as np
import yaml


def present_value(future_value, discount_rate, project_lifetime):
    """
    Calculate the present value of a future amount using the discount rate.
    """
    return future_value / ((1 + discount_rate) ** project_lifetime)

def calculate_lcoe(total_cost, total_generation, discount_rate, project_lifetime):
    """
    Calculate the Levelized Cost of Electricity (LCOE).
    """
    present_value_costs = present_value(total_cost, discount_rate, project_lifetime)
    present_value_generation = present_value(total_generation, discount_rate, project_lifetime)
    return present_value_costs / present_value_generation

# Load the configuration from a YAML file
yaml_file_path = "./inputs/test file 2-api alternative.yaml"
with open(yaml_file_path, 'r') as file:
    config = yaml.safe_load(file)

# Parameters initialization from the hybrid plant model
project_lifetime = 25  # years
discount_rate = 0.0588  # real discount rate, discount rate 8%, inflation rate 2%

# generation of each component over project lifetime
pv_total_generation = np.sum(hybrid_plant.generation_profile.pv)
wind_total_generation = np.sum(hybrid_plant.generation_profile.wind)
battery_total_generation = np.sum(hybrid_plant.generation_profile.battery)
genset_total_generation = np.sum(hybrid_plant.generation_profile.grid)  # Genset modeled as grid
total_system_generation_over_lifetime = pv_total_generation + wind_total_generation + genset_total_generation 

# Genset Specific Calculations
genset_capacity_kw = hybrid_plant.grid.interconnect_kw
genset_install_cost_per_kw = 500
replacement_cost_genset_per_kw = 500
om_cost_per_kw_per_op_hour = 0.03 # o&m cost in $/kw/op.hour
fuel_cost_per_l = 1.20
specific_fuel_consumption_l_per_kwh = 0.250
genset_operational_hours_per_year = np.sum(np.array(hybrid_plant.grid.generation_profile) > 0)/25 # Calculate the number of hours where generation is not zero per year
generator_operational_life_hours = 15000 # genset operational lifetime in hours
generator_operational_life_years = generator_operational_life_hours/genset_operational_hours_per_year # calculate genset operational life in years
num_genset_replacements = float(project_lifetime / generator_operational_life_years) - 1 # use decimal instead of a round up whole number
# Genset Costs
genset_installed_cost = genset_capacity_kw * genset_install_cost_per_kw
replacement_cost_genset = num_genset_replacements * genset_capacity_kw * replacement_cost_genset_per_kw
total_om_cost_genset = om_cost_per_kw_per_op_hour * genset_capacity_kw * genset_operational_hours_per_year * project_lifetime
annual_fuel_consumption = genset_total_generation / project_lifetime * specific_fuel_consumption_l_per_kwh
total_fuel_cost = annual_fuel_consumption * fuel_cost_per_l * project_lifetime
total_genset_cost = genset_installed_cost + replacement_cost_genset + total_om_cost_genset + total_fuel_cost

# Battery Costs
battery_capacity_kwh = config['technologies']['battery']['system_capacity_kwh']
battery_installed_cost = hybrid_plant.battery.total_installed_cost
om_cost_battery_per_kwh_per_year = 10  # O&M cost per kWh per year
operational_life_battery = 15  # Battery operational life in years
num_battery_replacements = float(project_lifetime/ operational_life_battery) - 1 # use decimal instead of a round up whole number
replacement_cost_battery_per_kWh = battery_installed_cost / battery_capacity_kwh  # Battery replacement cost, 700 $/kWh
replacement_cost_battery = num_battery_replacements * battery_capacity_kwh * replacement_cost_battery_per_kWh
total_om_cost_battery = om_cost_battery_per_kwh_per_year * battery_capacity_kwh * project_lifetime

total_battery_cost = battery_installed_cost + replacement_cost_battery + total_om_cost_battery

# PV Costs
pv_capacity_kw = hybrid_plant.system_capacity_kw.pv
pv_installed_cost = hybrid_plant.pv.total_installed_cost  # can be modified in hopp/tools/analysis/bos/cost_calculator.py
om_cost_pv_per_kw_per_year = 10  # PV O&M cost per kW per year
total_om_cost_pv = om_cost_pv_per_kw_per_year * pv_capacity_kw * project_lifetime
total_pv_cost = pv_installed_cost + total_om_cost_pv

# Wind Costs
wind_capacity_kw = hybrid_plant.system_capacity_kw.wind
wind_installed_cost = hybrid_plant.wind.total_installed_cost  # can be modified in hopp/tools/analysis/bos/cost_calculator.py
om_cost_wind_per_kw_per_year = 40  # Wind O&M cost per kW per year
total_om_cost_wind = om_cost_wind_per_kw_per_year * wind_capacity_kw * project_lifetime
total_wind_cost = wind_installed_cost + total_om_cost_wind

# Total System Costs 
total_system_cost = total_pv_cost + total_wind_cost + total_battery_cost + total_genset_cost

# total_load_served = np.sum(hybrid_plant.site.desired_schedule * 1000) * project_lifetime

# calculate total load served based on adjusted load
total_load_served = total_adjusted_demand * project_lifetime

# LCOE Calculation
lcoe = calculate_lcoe(total_system_cost, total_load_served, discount_rate, project_lifetime)
print(f"The calculated Levelized Cost of Electricity (LCOE) is: ${lcoe:.3f} per kWh")
```
