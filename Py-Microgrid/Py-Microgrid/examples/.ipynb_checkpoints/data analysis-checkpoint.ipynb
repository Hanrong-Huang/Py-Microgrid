```python
'''
Step 1: HOPP vs HOMER, where HOPP uses same system configuration as HOMER. Validate HOPPs consistency with HOMER. 
Each microgrid has the same system config (PV, wind, Battery, Genset) on HOMER and HOPP, both under Load Following dispatch
Disable flexible load for precise LCOE comparison

'''
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
auCopper1 = pd.read_csv('./simulation_results_HOPP_vs_HOMER_LF.csv', index_col=0, low_memory=False)  # HOPP LF
auCopper2 = pd.read_csv('./FL hybrid MGs + LF collection_v2.csv', index_col=0, low_memory=False)  # HOMER LF

# Create the plot
fig, ax = plt.subplots(figsize=(9, 6), dpi=300)

# Plot each dataset as a PDF (Probability Density Function) plot
sns.histplot(auCopper1['System LCOE ($/kWh)'], kde=True, bins=10, color='tab:green', label='HOPP Load Following', edgecolor=None)
sns.histplot(auCopper2['Cost/LCOE ($/kWh)'], kde=True, bins=10, color='tab:red', label='HOMER Load Following', edgecolor=None)



# Set the labels and title
plt.xlabel('LCOE ($/kWh)', fontsize=11)
plt.ylabel('Frequency', fontsize=11)

plt.grid(False)
# Add a legend
plt.legend(loc='best')

plt.savefig("Figure 1.png", dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```

```python
'''
Step 2: HOPP optimisation algorithm, showing results after using Nelder-Mead algorithm to minimise LCOE.
Now HOPP is optimised using Nelder-Mead optimisation algorithm to minimise LCOE
HOMER and HOPP are both under load following dispatch strategy
'''

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
auCopper1 = pd.read_csv('./parallel_simulations/simulation_results_modularised/combined_results_with_DR/combined_lowest_cost_systems_LF.csv', index_col=0, low_memory=False)  # HOPP LF after Nelder-Mead optimisation, without DR and flexible load
auCopper2 = pd.read_csv('./FL hybrid MGs + LF collection_v2.csv', index_col=0, low_memory=False)  # HOMER LF

# Create the plot
fig, ax = plt.subplots(figsize=(9, 6), dpi=300)

# Plot each dataset as a PDF (Probability Density Function) plot
sns.histplot(auCopper1['System LCOE ($/kWh)'], kde=True, bins=10, color='tab:green', label='HOPP Load Following + Nelder-Mead Optimisaiton', edgecolor=None)
sns.histplot(auCopper2['Cost/LCOE ($/kWh)'], kde=True, bins=10, color='tab:red', label='HOMER Load Following', edgecolor=None)



# Set the labels and title
plt.xlabel('LCOE ($/kWh)', fontsize=11)
plt.ylabel('Frequency', fontsize=11)
plt.grid(False)
# Add a legend
plt.legend(loc='best')

# Show the plot
plt.show()
```

```python
'''
Step 3: HOPP dispatch strategy optimisation
Compare HOPP LCOE distribution under different dispatch strategies. 
At this stage, DR and flexible load are disabled.
'''

## Plot LCOE distribution using violin plot for HOMER and HOPP (without DR and flexible load) 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Patch

# Load the datasets
auCopper1 = pd.read_csv('./parallel_simulations/simulation_results_modularised/combined_results_without_DR/FL hybrid MGs + LF collection_v2.csv', index_col=0, low_memory=False)  # HOMER hybrid microgrid
auCopper2 = pd.read_csv('./parallel_simulations/simulation_results_modularised/combined_results_without_DR/combined_lowest_cost_systems_PDR.csv', index_col=0, low_memory=False)  # HOPP Predictive DR
auCopper3 = pd.read_csv('./parallel_simulations/simulation_results_modularised/combined_results_without_DR/combined_lowest_cost_systems_LF.csv', index_col=0, low_memory=False)  # HOPP Heuristic Load Following
auCopper4 = pd.read_csv('./parallel_simulations/simulation_results_modularised/combined_results_without_DR/combined_lowest_cost_systems_heuristic.csv', index_col=0, low_memory=False)  # HOPP Heuristic Dispatch
auCopper5 = pd.read_csv('./parallel_simulations/simulation_results_modularised/combined_results_without_DR/combined_lowest_cost_systems_one_cycle_heuristic.csv', index_col=0, low_memory=False)  # HOPP One Cycle Heuristic

# Reset indices for each series to avoid alignment issues
auCopper1.reset_index(drop=True, inplace=True)
auCopper2.reset_index(drop=True, inplace=True)
auCopper3.reset_index(drop=True, inplace=True)
auCopper4.reset_index(drop=True, inplace=True)
auCopper5.reset_index(drop=True, inplace=True)

# Prepare data for the violin plot
data = pd.DataFrame({
    'HOMER Load Following': auCopper1['Cost/LCOE ($/kWh)'],
    'HOPP Predictive Dispatch': auCopper2['System LCOE ($/kWh)'],
    'HOPP Load Following': auCopper3['System LCOE ($/kWh)'],
    'HOPP Heuristic Dispatch': auCopper4['System LCOE ($/kWh)'],
    'HOPP One Cycle': auCopper5['System LCOE ($/kWh)']
})

# Melt the DataFrame for seaborn
melted_data = data.melt(var_name='Method', value_name='LCOE ($/kWh)')

# Create the violin plot with the specified colors
plt.figure(figsize=(12, 6), dpi=300)
palette = ['tab:red', 'tab:green', 'tab:blue', 'tab:purple', 'tab:orange']
sns.violinplot(data=melted_data, x='Method', y='LCOE ($/kWh)', palette=palette, inner='box')

# Create a custom legend using patches to match the violin plot colors
legend_patches = [Patch(color=palette[i], label=label) for i, label in enumerate(data.columns)]
plt.legend(handles=legend_patches, loc='upper right')

# Set the labels and title
plt.xlabel('', fontsize=11)
plt.ylabel('LCOE ($/kWh)', fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Show the plot
plt.show()
```

```python
'''
Step 3: HOPP dispatch strategy optimisation
Compare HOPP LCOE vs Renewable Fraction under different dispatch strategies. 
At this stage, DR and flexible load are disabled.
'''
## LCOE vs Renewable Fraction for HOPP 4 dispatch strategies
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import numpy as np
import matplotlib.patches as mpatches
from matplotlib.gridspec import GridSpec

# Load all datasets
homer_data = pd.read_csv('./parallel_simulations/simulation_results_modularised/combined_results_without_DR/FL hybrid MGs + LF collection_v2.csv', index_col=0, low_memory=False) # change to without DR 
auCopper1 = pd.read_csv('./parallel_simulations/simulation_results_modularised/combined_results_without_DR/combined_lowest_cost_systems_PDR.csv', index_col=0, low_memory=False)
auCopper2 = pd.read_csv('./parallel_simulations/simulation_results_modularised/combined_results_without_DR/combined_lowest_cost_systems_LF.csv', index_col=0, low_memory=False)
auCopper3 = pd.read_csv('./parallel_simulations/simulation_results_modularised/combined_results_without_DR/combined_lowest_cost_systems_heuristic.csv', index_col=0, low_memory=False)
auCopper4 = pd.read_csv('./parallel_simulations/simulation_results_modularised/combined_results_without_DR/combined_lowest_cost_systems_one_cycle_heuristic.csv', index_col=0, low_memory=False)

# Convert 'Ren Frac (%)' to numeric after stripping the '%' sign for all datasets
for df in [homer_data, auCopper1, auCopper2, auCopper3, auCopper4]:
    if 'Ren Frac (%)' in df.columns and df['Ren Frac (%)'].dtype == object:
        df['Ren Frac (%)'] = df['Ren Frac (%)'].str.rstrip('%').astype(float)

# Define the font size for labels
font_size = 11

# Create figure with custom GridSpec
fig = plt.figure(figsize=(14, 14), dpi=300)
gs = GridSpec(3, 2, height_ratios=[1, 1, 1])

# Create the HOMER subplot in the middle of first row
ax_homer = plt.subplot(gs[0, :1])  # Use just half the width
ax_homer.set_position([0.25, gs[0].get_position(fig).y0, 
                      0.5, gs[0].get_position(fig).height])  # Center it

# Create the 2x2 grid for HOPP plots
ax1 = plt.subplot(gs[1, 0])
ax2 = plt.subplot(gs[1, 1])
ax3 = plt.subplot(gs[2, 0])
ax4 = plt.subplot(gs[2, 1])

# Define plot information
plots = [
    ("HOMER Load Following", "Reds", homer_data, ax_homer, "a)", "Cost/LCOE ($/kWh)"),
    ("HOPP Predictive Dispatch", "Greens", auCopper1, ax1, "b)", "System LCOE ($/kWh)"),
    ("HOPP Load Following", "Blues", auCopper2, ax2, "c)", "System LCOE ($/kWh)"),
    ("HOPP Simple Battery", "Purples", auCopper3, ax3, "d)", "System LCOE ($/kWh)"),
    ("HOPP One Cycle", "Oranges", auCopper4, ax4, "e)", "System LCOE ($/kWh)")
]

# Create all plots
for label, cmap, data, ax, subplot_label, y_column in plots:
    plot = sns.histplot(
        data=data,
        x="Ren Frac (%)",
        y=y_column,
        bins=30,
        pmax=0.9,
        ax=ax,
        cbar=True,
        cmap=cmap
    )
    
    plot.collections[-1].colorbar.ax.set_ylabel('Number of Copper Deposits', fontsize=font_size)
    ax.set_xlabel("Renewable Fraction (%)", fontsize=font_size)
    ax.set_ylabel("LCOE ($/kWh)", fontsize=font_size)
    ax.text(-0.1, 1.05, subplot_label, transform=ax.transAxes, 
            fontsize=font_size, fontweight='bold', va='top', ha='right')
    
    # Create custom patch for legend
    patch = mpatches.Patch(color=sns.color_palette(cmap)[-2], label=label)
    ax.legend(handles=[patch], loc='upper right', frameon=True)
    
    # Set x-axis ticks at intervals of 5
    ax.xaxis.set_major_locator(mtick.MultipleLocator(5))
    ax.xaxis.set_major_formatter(mtick.StrMethodFormatter('{x:.0f}'))
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create figure and axes for 2x2 subplot
fig, axes = plt.subplots(2, 2, figsize=(16, 12), dpi=300)

# Load datasets with flexible load and DR
auCopper2_with = pd.read_csv('./parallel_simulations/simulation_results_modularised/combined_results_with_DR/combined_lowest_cost_systems_PDR.csv', index_col=0, low_memory=False)
auCopper3_with = pd.read_csv('./parallel_simulations/simulation_results_modularised/combined_results_with_DR/combined_lowest_cost_systems_LF.csv', index_col=0, low_memory=False)
auCopper4_with = pd.read_csv('./parallel_simulations/simulation_results_modularised/combined_results_with_DR/combined_lowest_cost_systems_heuristic.csv', index_col=0, low_memory=False)
auCopper5_with = pd.read_csv('./parallel_simulations/simulation_results_modularised/combined_results_with_DR/combined_lowest_cost_systems_one_cycle_heuristic.csv', index_col=0, low_memory=False)


# Load datasets without flexible load and DR
auCopper2_without = pd.read_csv('./parallel_simulations/simulation_results_modularised/combined_results_without_DR/combined_lowest_cost_systems_PDR.csv', index_col=0, low_memory=False)
auCopper3_without = pd.read_csv('./parallel_simulations/simulation_results_modularised/combined_results_without_DR/combined_lowest_cost_systems_LF.csv', index_col=0, low_memory=False)
auCopper4_without = pd.read_csv('./parallel_simulations/simulation_results_modularised/combined_results_without_DR/combined_lowest_cost_systems_heuristic.csv', index_col=0, low_memory=False)
auCopper5_without = pd.read_csv('./parallel_simulations/simulation_results_modularised/combined_results_without_DR/combined_lowest_cost_systems_one_cycle_heuristic.csv', index_col=0, low_memory=False)

# Define the datasets and their labels
data_pairs = [
    ('Predictive Dispatch', auCopper2_without, auCopper2_with, 'tab:green'),
    ('Load Following', auCopper3_without, auCopper3_with, 'tab:blue'),
    ('Heuristic Dispatch', auCopper4_without, auCopper4_with, 'tab:purple'),
    ('One Cycle', auCopper5_without, auCopper5_with, 'tab:orange')
]

# Plot labels
labels = ['a)', 'b)', 'c)', 'd)']

# Plot each strategy in its own subplot
for idx, (strategy, data_without, data_with, color) in enumerate(data_pairs):
    row = idx // 2
    col = idx % 2
    
    # Create histogram plots
    sns.histplot(
        data_with['System LCOE ($/kWh)'],
        kde=True,
        bins=10,
        color=color,
        alpha=0.3,
        label=f'{strategy}: With FL & DR',
        ax=axes[row, col],
        edgecolor='none'
    )
    sns.histplot(
        data_without['System LCOE ($/kWh)'],
        kde=True,
        bins=10,
        color='tab:red',  # Changed to red for all 'without DR' cases
        alpha=0.5,
        label=f'{strategy}: Without FL & DR',
        ax=axes[row, col],
        edgecolor='none'
    )
    
    
    # Add subplot label
    axes[row, col].text(-0.08, 1.02, labels[idx], 
                       transform=axes[row, col].transAxes, 
                       fontsize=11, 
                       fontweight='bold')
    
    # Customize each subplot
    axes[row, col].set_xlabel('LCOE ($/kWh)', fontsize=11)
    axes[row, col].set_ylabel('Frequency', fontsize=11)
    axes[row, col].legend(loc='upper right')
    axes[row, col].grid(False)
    
# Adjust layout to prevent overlap
#plt.tight_layout()
plt.show()
```
